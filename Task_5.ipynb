```python
# This step is important for understanding the diversity and distribution of values in numerical columns.
# It helps identify potential issues such as categorical variables misclassified as numerical or outliers.
# Additionally, it provides insights into the overall structure and variability of the dataset.


df_num = data.select_dtypes(include=np.number)


col_name = []
length = []


for col in df_num.columns:
    col_name.append(col)
    length.append(len(df_num[col].unique()))


df_2 = pd.DataFrame({'feature': col_name, 'count_of_unique_values': length})


df_2

```

```python
data.drop(['Description','Zipcode','Weather_Timestamp'],axis=1,inplace=True)
```

```python
# Plot boxplots for each numerical column


numerical_columns = data.select_dtypes(include='number').columns
for column in numerical_columns:
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.boxplot(data[column])
    ax.set(title=f'Boxplot for {column}', xlabel='Column', ylabel='Values')
    plt.show()
```

```python
# Plot boxplots for each numerical column without outliers
numerical_columns = data.select_dtypes(include='number').columns


for column in numerical_columns:
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.boxplot(data[column], showfliers=False)  # Remove outliers
    ax.set(title=f'Boxplot for {column}', xlabel='Column', ylabel='Values')
    plt.show()
```

```python
# This step is important for understanding the relationships between numerical features in the dataset.
# Correlation analysis helps identify potential patterns, dependencies, and multicollinearity among variables.
# It provides insights into which features are strongly correlated and may influence each other, 
# which is crucial for feature selection and predictive modeling.

plt.figure(figsize=(15, 9))
sns.heatmap(df_num.corr(), annot=True)
plt.show()
```

```python
# This step is crucial for identifying the cities with the highest number of accidents, 
# which can help prioritize resources for traffic management and safety measures. 
# Understanding the distribution of accidents across different cities is essential 
# for developing targeted interventions and policies to 
cities = data['City'].value_counts().head(10).index
top_cities_data = data[data['City'].isin(cities)]
city_counts = top_cities_data['City'].value_counts().reset_index()
city_counts.columns = ['City', 'Accidents Count']
fig = px.bar(city_counts, x='City', y='Accidents Count',
             labels={'City': 'City', 'Accidents Count': 'Accidents Count'},
             title='Number of Accidents in Top 10 Cities')
fig.show()
```

```python
accidents_severity = data.groupby('Severity').count()['ID']
accidents_severity
```

```python
#pie chart to show the severity distribution of accidents

fig, ax = plt.subplots(figsize=(8, 6), subplot_kw=dict(aspect="equal"))
labels = [1, 2, 3, 4]  # Assuming labels for severity levels
ax.pie(accidents_severity, labels=labels, autopct='%1.1f%%', pctdistance=0.85)

circle = plt.Circle((0, 0), 0.5, color='white')
ax.add_artist(circle)

ax.set_title("Accidents by Severity", fontdict={'fontsize': 16})

plt.tight_layout()
plt.show()
```

```python


weather_counts = data['Weather_Condition'].value_counts()
top10_weather_counts = weather_counts.head(5)

plt.figure(figsize=(8, 6))
plt.pie(top10_weather_counts, labels=top10_weather_counts.index, autopct='%1.1f%%')
plt.title('Top 5 Weather Conditions Distribution')
plt.show()
```

```python
#show Accidents Count By Time of Day

data['Start_Time'] = pd.to_datetime(data['Start_Time'], errors='coerce')
data = data.dropna(subset=['Start_Time'])

plt.figure(figsize=(8, 5))
sns.histplot(data['Start_Time'].dt.hour, bins=24)

plt.xlabel("Start Time")
plt.ylabel("Number of Occurrences")
plt.title('Accidents Count By Time of Day')

plt.show()
```

```python
#show Accidents Count By Time of Day

data['End_Time'] = pd.to_datetime(data['End_Time'], errors='coerce')
data = data.dropna(subset=['End_Time'])

plt.figure(figsize=(8, 5))
sns.histplot(data['End_Time'].dt.hour, bins=24)

plt.xlabel("End Time")
plt.ylabel("Number of Occurrences")
plt.title('Accidents Count By Time of Day')

plt.show()
```
